```python
df = pd.read_csv('world_carona_data.csv')
df
```

# Filling null values

```python
df.loc[:,df.columns] = df.loc[:,df.columns].fillna(value=0)
```

# Country wise created reports

```python
temp_f = df.sort_values(by='TotalCases', ascending=False)
temp_f = temp_f[['Country,Other', 'TotalCases', 'ActiveCases', 'NewDeaths', 'TotalRecovered']]
temp_f = temp_f.reset_index(drop=True)

temp_f.style.background_gradient(cmap="Blues", subset=['TotalCases', 'ActiveCases'])\
            .background_gradient(cmap="Greens", subset=['TotalRecovered'])\
            .background_gradient(cmap="Reds", subset=['NewDeaths'])
```

# Donut plot showing total Active,Recovered and Deaths

```python
#Overall 
ac= df['ActiveCases'].sum()
rvd = df['TotalRecovered'].sum()
dth = df['TotalDeaths'].sum()
fig = go.Figure(data=[go.Pie(labels=['Active Cases','Cured','Deaths'],
                             values= [ac,rvd,dth],hole =.3)])
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=20,
                  marker=dict(colors=['#263fa3', '#2fcc41','#cc3c2f'], line=dict(color='#FFFFFF', width=2)))
fig.show()
```

# Pie plot showing Active cases w.r.t Countries

```python
fig = px.pie(df, values = 'ActiveCases',names='Country,Other', height=600)
fig.update_traces(textposition='inside', textinfo='percent+label')

fig.update_layout(
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
    ))

fig.show()
```

# Tree Map shows Country having Total cases

```python
fig = px.treemap(df, path=['Country,Other'], values='TotalCases', height=600, width=1000)

fig.update_layout(
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
    ))

fig.show()
```

# Funnel area chart shows Country wise total deaths

```python
fig = px.funnel_area(df,names='Country,Other', values='TotalDeaths' ,height=600 )
fig.show()
```

# Geographical globe impact on covid-19

```python
total_df = df.groupby(['Country,Other','TotalCases'])['TotalDeaths','TotalRecovered','Serious,Critical'].sum()
total_df = total_df.reset_index()
fig = px.scatter_geo(total_df,locations="Country,Other", locationmode='country names', 
                      color="TotalCases", size='TotalRecovered',hover_name="Country,Other", lon='TotalDeaths',lat='Serious,Critical',
                      projection="orthographic",color_continuous_scale=px.colors.sequential.deep,
                      )
fig.update_layout(
    title='Total Details over Countries',
)
fig.show()
```

```python
total_df = df.groupby(['Country,Other','TotalCases'])['TotalDeaths','TotalRecovered','Serious,Critical'].sum()
total_df = total_df.reset_index()
fig = px.scatter_geo(total_df,locations="Country,Other", locationmode='country names', 
                      color="TotalCases", size='TotalRecovered',hover_name="Country,Other", lon='TotalDeaths',lat='Serious,Critical',
                      projection="natural earth",
                      )
fig.update_layout(
    title='Total Details over Countries',
)
fig.show()
```

```python
total_df = df.groupby(['Country,Other','TotalCases'])['TotalDeaths','TotalRecovered'].sum()
total_df = total_df.reset_index()
fig = px.scatter_geo(total_df, locations="Country,Other", locationmode='country names', 
                      color="TotalRecovered", size='TotalCases', hover_name="Country,Other", 
                      projection="natural earth", animation_frame="TotalDeaths", 
                      title='Total Details over Death Cases'
                      )
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

# World map showing confirmed cases of covid-19

```python
Data_per_country = df.groupby(["Country,Other"])["TotalCases"].sum().reset_index().sort_values("TotalCases",ascending=False).reset_index(drop=True)
fig = px.choropleth(Data_per_country, locations=Data_per_country['Country,Other'],
                    color=Data_per_country['TotalCases'],locationmode='country names', 
                    hover_name=Data_per_country['Country,Other'], 
                    color_continuous_scale=px.colors.sequential.deep)
fig.update_layout(
    title='Confirmed Cases In Each Country',
)
fig.show()
```

# World map showing Recovered cases of covid-19

```python
Recovered_per_country = df.groupby(["Country,Other"])["TotalRecovered"].sum().reset_index().sort_values("TotalRecovered",ascending=False).reset_index(drop=True)
fig = px.choropleth(Recovered_per_country, locations=Recovered_per_country['Country,Other'],
                    color=Recovered_per_country['TotalRecovered'],locationmode='country names', 
                    hover_name=Recovered_per_country['Country,Other'], 
                    color_continuous_scale=px.colors.sequential.deep)
fig.update_layout(
    title='Recovered Cases In Each Country',
)
fig.show()
```

# World map showing Recovered cases of covid-19

```python
Recovered_per_country = df.groupby(["Country,Other"])["TotalDeaths"].sum().reset_index().sort_values("TotalDeaths",ascending=False).reset_index(drop=True)
fig = px.choropleth(Recovered_per_country, locations=Recovered_per_country['Country,Other'],
                    color=Recovered_per_country['TotalDeaths'],locationmode='country names', 
                    hover_name=Recovered_per_country['Country,Other'], 
                    color_continuous_scale=px.colors.sequential.deep)
fig.update_layout(
    title='Recovered Cases In Each Country',
)
fig.show()
```
